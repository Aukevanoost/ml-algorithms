# Derivatives

There are a couple of rules for derivatives: 

`α` = constant
1. `der(x): α` = `0`
2. `der(x): α * xⁿ` = `α * n * xⁿ⁻¹`
3. `der(x): expA + expB` = `der(x): expA + der(x): expB`
4. `der: α * expC` = `α * der(x): expC`


Chain rule:
5. `der: f(g(x)) = der(g(x)): f(g(x)) * der(x): g(x)`

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
graphs = [
    # Functions                                 # Derivatives
    (lambda x: 2*x + 3,                         lambda _: 2                      ),
    (lambda x: x ** 2,                          lambda x: 2 * x                  ),
    (lambda x: 3 * (x**3) + 2 * (x**2) - x - 2, lambda x: 9 * x ** 2 + 4 * x - 1 ),
    (lambda x: (x ** 2 + 1) ** 2,               lambda x: (4 * x**3) + 4 * x     )
]

_, axes = plt.subplots(nrows=1,ncols=len(graphs), figsize=(20, 5))
for i in range(len(graphs)):
    (f, der) = graphs[i]
    df = pd.DataFrame([[ f(x), der(x) ] for x in range(-10, 11)], columns=["fn(x)", "derivative"])
    df.plot(kind='line', ax=axes[i])
```

## First function: 2*x + 3

**Rules**
1. `der(x): α` = `0`
2. `der(x): α * xⁿ` = `α * n * xⁿ⁻¹`
3. `der(x): expA + expB` = `der(x): expA + der(x): expB`

**Application**
1. Applying rule 3: `2*x + 3` becomes `2*x` + `3`
2. Applying rule 2: (2 is a α)`
    - `2 * x` is actually `2 * x¹` so becomes `2 * 1 * x¹⁻¹`
    - `2*n*x¹⁻¹` is normalized `2 * 1 * x⁰`
    - `2 * n * x⁰` is normalized `2 * 1 * 1`
    - `2 * 1 * 1` is normalized `2`
3. Applying rule 1: `3` is a constant (α) so becomes `0`
4. Normalized `2 + 0` becomes `2`


Output: `der(x): 2`

## Second function: x ^ 2

**Rules**
1. `der(x): α` = `0`
2. `der(x): α * xⁿ` = `α * n * xⁿ⁻¹`
3. `der(x): expA + expB` = `der(x): expA + der(x): expB`

**Application**
1. Applying rule 2: `x²` becomes `1 * 2 * x`

Output: `der(x): 2 * x`

## Third function: 3 * x³ + 2 * x² - x - 2

**Rules**
1. `der(x): α` = `0`
2. `der(x): α * xⁿ` = `α * n * xⁿ⁻¹`
3. `der(x): expA + expB` = `der(x): expA + der(x): expB`

**Application**
1. Applying rule 3: breaking up in `3*x³`, `2 * x²`, `-x` and `-2`
2. Applying rule 2: `3 * x³` becomes `3 * 3 * x²` which is normalized `9 * x²`
3. Applying rule 2: `2 * x²` becomes `2 * 2 * x¹` which is normalized `4 * x`
4. Applying rule 2: `-x` becomes `1 * 1 * -x⁰` which is normalized `-1`
5. Applying rule 1: `-2` becomes `0`
5. Combined is : `9 * x² + 4 * x + -1 + 0` which is normalized `9 * x² + 4 * x - 1`

Output: `9 * x² + 4 * x - 1`

## Fourth function: (x² + 1)²

**Rules**
1. `der(x): α` = `0`
2. `der(x): α * xⁿ` = `α * n * xⁿ⁻¹`
3. `der(x): expA + expB` = `der(x): expA + der(x): expB`
4. `der(x): α * expC` = `α * der(x): expC`
5. `der(x): f(g(x)) = der(g(x)): f(g(x)) * der(x): g(x)`

**Application**
1. Applying rule 5 to `(x² + 1)²`:
    - f is a composite fn: `( g(x) )²` where `g(x) = x² + 1`
    - applying rule 5: `(x² + 1)²` becomes `(g(x)²)` * `(x² + 1)`
2. Processing the first part: `(g(x)²)`
    - Applying rule 2 to f: `g(x)²` is `2 * g(x)`
    - defining g(x): `2 * (x² + 1)`
2. Processing the second part: (`x² + 1`)
    - Applying rule 3: `x² + 1` is `x²` and `1`.
    - Applying rule 1: `1` is `0` is combined `x² + 0` is `x²`
    - Applying rule 2: `x²` is `1 * 2 * x¹` is `2x`
3. Merging the parts: `(2 * (x² + 1))` * `2 * x`
    - math rule (a + b) * c becomes c * a + c * b: `(2 * x² + 2 * 1)` * `2 * x`
    - is formula: `(2x² + 2) * 2x`
    - math rule (a + b) * c becomes c * a + c * b: `2x * 2x²` + `2x * 2`
    - part A is: `2x * 2x²` is `2 * x * 2 * x²` is `4 * x * x²` is `4x³`
    - part B is: `2x * 2` is `2 * x * 2` is `2 * 2 * x` is `4x`
    - final formula: `4x³ + 4x`

Output: `4x³ + 4x`
